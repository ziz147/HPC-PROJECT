# Calcul des courbes de speed-up et d'efficacité

## Speed-up

Le speed-up est calculé en comparant le temps d'execution du programme $T(1)$ sur 1 processeur à celui sur $p$ processeurs $T(p)$:

$S(p) = \frac{T(1)}{T(p)}$

## Efficacité

Elle est définie comme le rapport du speed-up au nombre de processeurs utilisés:

$E(p) = \frac{S(p)}{p} = \frac{T(1)}{p \times T(p)}$

où $E(p)$ est l'efficacité avec $p$ processeurs.

```python
import numpy as np
import matplotlib.pyplot as plt


# Remplacez ceci par le chemin de votre fichier de données
chemin_fichier =  'execution_times.txt'

# Lire les données du fichier
temps_execution = []
with open(chemin_fichier, 'r') as file:
    for line in file:
        # Assurez-vous que le format de vos temps d'exécution dans le fichier est cohérent
        # et modifiez la conversion ci-dessous en conséquence.
        temps_execution.append(float(line.strip()))
TE = np.array(temps_execution)

# Assurez-vous que le premier élément est le temps d'exécution pour un seul processeur
T1 = temps_execution[0]

# Calculer le speed-up et l'efficacité
speed_up = T1 / TE
efficacite = T1 / (np.arange(1, TE.shape[0]+1, 1) * TE)

# Tracer le temps
plt.figure(3)
plt.plot(range(1, len(speed_up) + 1), TE, marker='o')
plt.title('Temps d\'execution')
plt.xlabel('Nombre de processeurs')
plt.ylabel('Temps d\'execution')
plt.grid(True)

# Tracer le speed-up
plt.figure(1)
plt.plot(range(1, len(speed_up) + 1), speed_up, marker='o')
plt.plot(range(1, len(speed_up) + 1), range(1, len(speed_up) + 1), marker='*')
plt.title('Speed-up')
plt.xlabel('Nombre de processeurs')
plt.ylabel('Speed-up')
plt.grid(True)

# Tracer l'efficacité
plt.figure(2)
plt.plot(range(1, len(efficacite) + 1), efficacite, marker='x', color='r')
plt.title('Efficacité')
plt.xlabel('Nombre de processeurs')
plt.ylabel('Efficacité')
#plt.ylim(0, 1)  # L'efficacité est généralement entre 0 et 1
plt.grid(True)

plt.show()
```
