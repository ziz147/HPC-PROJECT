# Calcul des courbes de speed-up et d'efficacité

## Speed-up

Le speed-up est calculé en comparant le temps d'execution du programme $T(1)$ sur 1 processeur à celui sur $p$ processeurs $T(p)$:

$S(p) = \frac{T(1)}{T(p)}$

## Efficacité

Elle est définie comme le rapport du speed-up au nombre de processeurs utilisés:

$E(p) = \frac{S(p)}{p} = \frac{T(1)}{p \times T(p)}$

où $E(p)$ est l'efficacité avec $p$ processeurs.

```python
import numpy as np
import matplotlib.pyplot as plt


# Remplacez ceci par le chemin de votre fichier de données
chemin_fichier =  'execution_times.txt'
fichier_plafrim = 'plafrim_local_3d.txt'

# Lire les données du fichier
temps_execution = []
temps_execution_plafrim = []

with open(fichier_plafrim, 'r') as file:
    for line in file:
        # Assurez-vous que le format de vos temps d'exécution dans le fichier est cohérent
        # et modifiez la conversion ci-dessous en conséquence.
        temps_execution_plafrim.append(float(line.strip()))
        #temps_execution_plafrim.append()
TEP = np.array(temps_execution_plafrim)
with open(chemin_fichier, 'r') as file:
    for line in file:
        # Assurez-vous que le format de vos temps d'exécution dans le fichier est cohérent
        # et modifiez la conversion ci-dessous en conséquence.
        temps_execution.append(float(line.strip()))
        #temps_execution_plafrim.append()
TE = np.array(temps_execution)

# Assurez-vous que le premier élément est le temps d'exécution pour un seul processeur
T1 = temps_execution[0]
T2 = temps_execution_plafrim[0]

# Calculer le speed-up et l'efficacité
speed_up = T1 / TE
speed_up_plafrim = T2/TEP
efficacite = T1 / (np.arange(1, TE.shape[0]+1, 1) * TE)
efficacitep = T2 / (np.arange(1, TEP.shape[0]+1, 1) * TEP)

# Tracer le temps
plt.figure(3)
plt.plot(range(1, len(speed_up) + 1), TE, marker='o', label ='Using local machine')
plt.plot(range(1, len(speed_up_plafrim) + 1), TEP, marker='o', label='Using plafrim')
plt.title('Execution time in 3D')
plt.xlabel('Number of processors')
plt.ylabel('Execution time')
plt.legend()
plt.grid(True)

# Tracer le speed-up
plt.figure(1)
plt.plot(range(1, len(speed_up) + 1), speed_up, marker='o', label ='Using local machine')
plt.plot(range(1, len(speed_up_plafrim) + 1), speed_up_plafrim, marker='*',label='Using plafrim' )
plt.plot(range(1, len(speed_up_plafrim) + 1), range(1, len(speed_up_plafrim) + 1), label = 'Ideal speedup')
plt.title('Speed-up in 3D')
plt.xlabel('Number of processors')
plt.ylabel('Speed-up')
plt.legend()
plt.grid(True)

# Tracer l'efficacité
plt.figure(2)
plt.plot(range(1, len(efficacite) + 1), efficacite, marker='x', color='r', label = 'Using local machine')
plt.plot(range(1, len(efficacitep) + 1), efficacitep, marker='x', color='g', label = 'Using Plafrim')
plt.title('efficiency in 3D ')
plt.xlabel('Number of processors')
plt.ylabel('efficiency')
#plt.ylim(0, 1)  # L'efficacité est généralement entre 0 et 1
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Remplacer ces listes par vos données de temps d'exécution réelles pour les deux matrices
temps_python_A = [1.2, 1.3, 1.1]  # Exemple de temps d'exécution pour Python avec la matrice A
temps_c_A = [0.8, 0.7, 0.75]      # Exemple de temps d'exécution pour C avec la matrice A
temps_c_parallel_A = [0.4, 0.5, 0.45]  # Exemple pour C en parallèle avec la matrice A

temps_python_B = [1.5, 1.4, 1.6]  # Exemple de temps d'exécution pour Python avec la matrice B
temps_c_B = [0.9, 0.85, 0.8]      # Exemple de temps d'exécution pour C avec la matrice B
temps_c_parallel_B = [0.45, 0.5, 0.55]  # Exemple pour C en parallèle avec la matrice B

# Calculer les moyennes
moyennes_A = [np.mean(temps_python_A), np.mean(temps_c_A), np.mean(temps_c_parallel_A)]
moyennes_B = [np.mean(temps_python_B), np.mean(temps_c_B), np.mean(temps_c_parallel_B)]

# Paramètres de l'histogramme
barWidth = 0.35
r1 = np.arange(len(moyennes_A))
r2 = [x + barWidth for x in r1]

# Création de l'histogramme
plt.bar(r1, moyennes_A, color='blue', width=barWidth, edgecolor='grey', label='Matrice A')
plt.bar(r2, moyennes_B, color='green', width=barWidth, edgecolor='grey', label='Matrice B')

# Ajout de légendes et titres
plt.xlabel('Méthode', fontweight='bold')
plt.xticks([r + barWidth / 2 for r in range(len(moyennes_A))], ['Python', 'C', 'C Parallèle'])
plt.ylabel('Temps d\'Exécution Moyen (s)')
plt.title('Comparaison des Temps d\'Exécution pour les Matrices A et B')

plt.legend()

# Afficher le graphique
plt.show()
```
